# Which Neighbourhood takes AirBNB frequently

```python
fig, ax = plt.subplots(figsize=(80,78))
width = len(df1['neighbourhood_group'].unique()) + 1
fig.set_size_inches(width , 8)
ax=sns.countplot(data = df1, x= 'neighbourhood_group')



for p in ax.patches:
    ax.annotate(str((np.round(p.get_height()/len(df1)*100,decimals=2)))+'%', (p.get_x()+p.get_width()/2., p.get_height()), ha='center', va='center', xytext=(0, 10), textcoords='offset points')
```
